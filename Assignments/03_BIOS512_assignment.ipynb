Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
theme_set(theme_light(base_size = 12))

options(repr.plot.width = 7, repr.plot.height = 6)

p = ggplot(penguins, aes(x = body_mass_g)) +
    geom_histogram(breaks = seq(2650, 6350, 100), na.rm = TRUE, alpha = 0.75, binwidth = 70, color="steelblue", fill="steelblue") +
    facet_wrap(~year, ncol = 1) +
    labs(title = 'Histograms of penguin body mass for measurements taken from 2007-2009', y = 'Number of penguins', x = 'Body mass in grams') +
    theme_minimal() +
    theme(strip.text.x = element_text(hjust = 0.9175), plot.title = element_text(hjust = -0.16, vjust = 3, size = 9)) +
    scale_x_continuous(labels = scales::comma_format())
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width = 5, repr.plot.height = 6)

p = ggplot(penguins, aes(x = body_mass_g, y = sex)) + 
    facet_wrap(~species, ncol = 1) +
    geom_point(na.rm = TRUE, position = position_jitter(h = 0.2, w = 0), size = 1) +
    labs(x = 'Body mass in grams', y = '') +
    theme_minimal() +
    theme(strip.text.x = element_text(hjust = 0.08)) + 
    scale_x_continuous(breaks = c(3500, 4500, 5500), labels = scales::comma_format())
p
```

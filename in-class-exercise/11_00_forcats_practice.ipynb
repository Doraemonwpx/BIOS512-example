Make a bar chart of the frequency of values for one "categorical" column in the data you chose.

```r
volcano = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-12/volcano.csv',
                  col_types = cols())
volcano2 = volcano |>
    select(volcano_name, region, last_eruption_year) |>
    mutate(region = region |> fct_lump(n = 8) |> fct_infreq() |> fct_rev() |> fct_relevel('Other', after = Inf))
p = ggplot(volcano2, aes(y = region)) + geom_bar()
p
```
